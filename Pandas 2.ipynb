Multi Index

```python
# Creating multi index
#index level
outside='Jakarta Jakarta Jakarta Surabaya Surabaya Surabaya'.split()
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
hier_index
```

```python
hier_index=pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df_multi=pd.DataFrame(np.random.randint(1,100,12).reshape(6,2), index=hier_index, columns='Smartphone PC'.split())
df_multi
```

```python
df_multi.loc['Jakarta'] #MultiIndex memudahkan kita mencari data dalam kelompok tertentu
# menampilkan data penjualan di seluruh toko di Jakarta
```

```python
df_multi.loc['Jakarta'].loc[2] # menampilkan data penjualan toko 2 di kota jakarta
```

```python
df_multi.xs(['Jakarta',2]) #selain dengan .loc, kita bisa memanggil data multi index menggunakan .xs
```

```python
df_multi
```

```python
df_multi.index.names #index df_multi belum memiliki nama
```

```python
df_multi.index.names=['Kota','Lokasi'] #memberi nama pada index
df_multi
```

```python
df_multi.xs(1,level='Lokasi') #mengakses data multi index menggunakan nama index
# toko-toko di lokasi 1 dari setiap kota
```

```python
df_multi.xs('Jakarta')
```

```python
df_multi.iloc[0]
```

```python
df_multi.loc['Jakarta']
```
