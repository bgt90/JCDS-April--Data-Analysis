# Merge and Join

```python
import pandas as pd
import numpy as np
```

```python
left=pd.DataFrame({'key':'K0 K1 K2 K3'.split(),
                 'A':'A0 A1 A2 A3'.split(),
                 'B': 'B0 B1 B2 B3'.split()})

right=pd.DataFrame({'key':'K0 K1 K3 K4'.split(),
                 'C':'C0 C1 C3 C4'.split(),
                 'D': 'D0 D1 D3 D4'.split()})
```

```python
left
```

```python
right
```

**Merge**

```python
pd.merge(left,right,on='key') #dataframe kiri, dataframe kanan, on='nama kolom yang menjadi key'
#secara default, inner join
```

```python
pd.merge(left,right,how='left', on='key')
#how= cara join: inner, left, right, outer
# di sini kita coba left join, sehingga key yang dipakai mengikuti key yang ada di dataframe left
```

```python
pd.merge(left,right,how='right', on='key')
#right join, key mengikuti dataframe right
```

```python
pd.merge(left,right,how='outer', on='key')
#outer join, seua key dari kedua dataframe muncul
```

**Join**  
mirip merge, tapi digabungkan berdasarkan index, bukan kolom tertentu

```python
left=left.set_index('key')
right=right.set_index('key')
```

```python
left
```

```python
right
```

```python
left.join(right) #default left join
```

```python
left.join(right, how='inner')
```

```python
left.join(right, how='outer')
```

```python
left=pd.DataFrame({'key':'K0 K1 K2 K3'.split(),
                 'A':'A0 A1 A2 A3'.split(),
                 'B': 'B0 B1 B2 B3'.split()})

right=pd.DataFrame({'key':'K0 K1 K3 K4'.split(),
                 'B': 'B0 B1 B2 B3'.split(),
                 'D': 'D0 D1 D3 D4'.split()})
```

```python
left
```

```python
left=left.set_index('key')
right=right.set_index('key')
```

# Concatenating

```python
df1=pd.DataFrame({
    'A':'A0 A1 A2 A3'.split(),
    'B':'B0 B1 B2 B3'.split(),
    'C':'C0 C1 C2 C3'.split(),
    'D':'D0 D1 D2 D3'.split()
})

df2=pd.DataFrame({
    'A':'A4 A5 A6 A7'.split(),
    'B':'B4 B5 B6 B7'.split(),
    'C':'C4 C5 C6 C7'.split(),
    'D':'D4 D5 D6 D7'.split()
})
df3=pd.DataFrame({
    'A':'A8 A9 A10 A11'.split(),
    'B':'B8 B9 B10 B11'.split(),
    'C':'C8 C9 C10 C11'.split(),
    'D':'D8 D9 D10 D11'.split()
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2,df3]) #menempel dataframe, secara default berdasarkan baris
```

```python
pd.concat([df1,df2,df3],axis=1) #digabung ke samping
```

# Apply

```python
df=pd.DataFrame({
    'col1':[1,2,3,4],
    'col2':[444,555,666,444],
    'col3':'abc def ghi xyz'.split()
})
df
```

```python
def times2(x):
    return x*2
df['col1'].apply(times2) #apply digunakan untuk mengaplikan suatu function terhadap tiap baris di suatu kolom
```

```python
df['col1'].apply(lambda x:x*2) # bisa juga pakai lambda sebagai pengganti function times2
```

```python
len(df['col3']) #panjang dari series df['col3'], padahal ingin mencari panjang kata dari tiap baris di df['col3']
```

```python
df['col3'].apply(len)
```

```python
df['col3'].apply(lambda x: len(x)*2) #contoh pemakaian lambda apply
```

```python
df['col3'].apply(lambda x: x[0]) #mengambil huruf pertama dari tiap baris di col1
#apply memang lebih sering dipakai dalam pengolahan string di pandas
```

# Pivot Table

```python
data= pd.DataFrame({
    'Kota':'Jakarta Jakarta Jakarta Surabaya Surabaya Surabaya'.split(),
    'Toko':'one one two two one one'.split(),
    'Manager':'Andi Budi Andi Budi Andi Budi'.split(),
    'Penjualan':[1,3,2,5,4,1]
})
data
```

```python
data.pivot_table(values='Penjualan', index=['Kota','Toko'],columns='Manager')
```

```python
data.loc[6]=['Surabaya','one','Budi',4]
data
```

```python
data.pivot_table(values='Penjualan', index=['Kota','Toko'],columns='Manager') #secara default, akan mencari rata-rata
```

```python
df=pd.DataFrame({
    'date':[1,2,3,1,2,3,1,2,3,1,2,3],
    'variabel':'A A A B B B C C C D D D'.split(),
    'value': list(np.random.randint(1,50,12))
})
df
```

```python
df.pivot(index='date',columns='variabel',values='value') #mengubah struktur data
```
